```python
import xarray as xr
from dask.distributed import Client
```

```python
mesh = xr.open_dataset('/home/sallen/MEOPAR/grid/mesh_mask202108.nc')
# the mask is 1 where there is water, we want the opposite.  The meshmask has an extra dimension, hence the [0]
tmask = 1 - mesh.tmask[0]

volume_def = ((1-tmask) * mesh.e3t_0[0] * mesh.e1t[0] * mesh.e2t[0]).rename({'z':'depth', 'y':'gridY', 'x':'gridX'})
rho_inv = (1/1026)

# e3t_1d = mesh.e3t_1d.rename({'z':'depth'}).isel(t=0)
e3t_3d = mesh.e3t_0.rename({'z':'depth', 'y':'gridY', 'x':'gridX'}).isel(t=0)
```

```python
chunk_sizes = {'time': 1, 'depth': 40, 'gridY': 898, 'gridX': 398}
```

```python
year = 2019
fname_vol = 'SalishSeaCast_river_trace_day_avg_allR_00_40_{}0101_{}1231_e3t_3d_vol'.format(year, year)
fname_depsum = 'SalishSeaCast_river_trace_day_avg_allR_00_40_{}0101_{}1231_e3t_3d_vol_e3t_depsum_vol'.format(year, year)
volume = xr.open_dataset('/ocean/cdonaldson/research_runs/hindcast_auto/transformations/3d_hindcast_vol/{}.nc'.format(fname_vol),chunks=chunk_sizes)
depsum = xr.open_dataset('/ocean/cdonaldson/research_runs/hindcast_auto/transformations/3d_hindcast_vol/{}.nc'.format(fname_depsum),chunks=chunk_sizes)
```

```python
volume  # cubic meters of dye
```

```python
# e3t_1d = mesh.e3t_1d.rename({'z':'depth'}).isel(t=0)
e3t_3d = mesh.e3t_0.rename({'z':'depth', 'y':'gridY', 'x':'gridX'}).isel(t=0)
```

```python
depsum
```

```python
client=Client(processes=True, n_workers=4, local_directory='/tmp/cdonaldson')  # 4 is a good choice for n_workers even on salish, if I was the only person logged on or asked for space, could go as high as 16 but... be conscious of this
# threads_per_worker = 1 if needed, netcdf is not threadsafe now so be realiable by specifying 1 (at expense of just some efficiency)
```

```python
avg_depth = (volume * e3t_3d).sum(dim='depth') / depsum
```

```python
for data_var in list(avg_depth.data_vars):
    avg_depth[data_var].attrs['units'] = 'm^3'
```

```python
fname_save = fname_vol[:-10] + 'avg_depth'
fname_save
```

```python
avg_depth.to_netcdf('/ocean/cdonaldson/research_runs/hindcast_auto/transformations/3d_hindcast_vol/{}.nc'.format(fname_save), engine="netcdf4")
```
