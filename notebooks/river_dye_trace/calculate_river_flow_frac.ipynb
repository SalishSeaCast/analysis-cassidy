```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
river_flows_puget = pd.read_csv('/ocean/cdonaldson/MOAD/analysis-cassidy/notebooks/river_dye_trace/river_dailies_to_vol_ts_all_puget_20190101_20231231.csv')
river_flows_puget = river_flows_puget.loc[:, river_flows_puget.columns != 'Unnamed: 0']

river_flows_puget['date'] = pd.to_datetime(river_flows_puget['date'])
river_flows_puget['month'] = river_flows_puget['date'].dt.month
river_flows_puget['dayofyear'] = river_flows_puget['date'].dt.dayofyear
river_flows_puget['year'] = river_flows_puget['date'].dt.year
plt.plot(river_flows_puget['date'], river_flows_puget['Johnson [m^3/s]'].values)
```

```python
river_flows_skagit = pd.read_csv('/ocean/cdonaldson/MOAD/analysis-cassidy/notebooks/river_dye_trace/river_dailies_to_vol_ts_all_skagit_20190101_20231231.csv')
river_flows_skagit = river_flows_skagit.loc[:, river_flows_skagit.columns != 'Unnamed: 0']

river_flows_skagit['date'] = pd.to_datetime(river_flows_skagit['date'])
river_flows_skagit['month'] = river_flows_skagit['date'].dt.month
river_flows_skagit['dayofyear'] = river_flows_skagit['date'].dt.dayofyear
river_flows_skagit['year'] = river_flows_skagit['date'].dt.year
plt.plot(river_flows_skagit['date'], river_flows_skagit['Mission [m^3/s]'].values)
```

```python
just_flows_puget = river_flows_puget.loc[:, ~river_flows_puget.columns.isin(['filename', 'date', 'month', 'dayofyear', 'year'])]
just_flows_puget
```

```python
just_flows_skagit = river_flows_skagit.loc[:, ~river_flows_skagit.columns.isin(['filename', 'date', 'month', 'dayofyear', 'year'])]
just_flows_skagit
```

```python
just_flows_ps = pd.concat([just_flows_puget, just_flows_skagit], axis=1)
just_flows_ps
```

```python
flows_daily_average = just_flows_ps.sum() / len(just_flows_ps)
sorted_daily_avg = flows_daily_average.sort_values(ascending=False)
sorted_daily_avg
```

```python
prop = sorted_daily_avg / sorted_daily_avg.sum()
prop.plot.bar()
```
